```python
# Using cube of first 1000 numbers using for loop
for i in range(1,1001):
     cube = i ** 3
     print(cube)
```

```python
num = 0
while num <= 1000:
    cube = num ** 3
    print(f"The cube of {num} is " + str(cube))
    num +=1
   
```

```python
# Prime numbers up to 100
for  i in range(1,101):
    if i > 1 :
        for j in range(2,i): 
            if (i%j==0):
               break
        else: 
            print(i)
```
